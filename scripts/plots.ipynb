```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
batch_list = [10,100,1000,10000,100000]
batch_array = np.array(batch_list)
inf_time_list = [56.66,45.79,67.57,94.76,243.025]
inf_time_array = np.array(inf_time_list)*1000
time_per_inf = np.divide(inf_time_array,batch_array)
actual_inf_time = 0.113*1000*np.ones(5)
fig,ax = plt.subplots()
ax.loglog(batch_array,time_per_inf,label='AMPNet Inference Time')
ax.loglog(batch_array,actual_inf_time,label = 'Reed Shepps Evaluation Time')
plt.grid(True)
plt.xlabel('Batch Size')
plt.ylabel('Time per Inference(us)')
plt.legend()
plt.savefig('infer_time.png')
plt.show()
```
